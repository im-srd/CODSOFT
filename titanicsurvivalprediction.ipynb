 # Loading the Data
 
 * **train data** -> train.csv
 * **test data** -> tested.csv

```python
trainDataTit = pd.read_csv("/kaggle/input/titanic/train.csv")
print(trainDataTit.head)
```

```python
testDataTit = pd.read_csv("/kaggle/input/test-file/tested.csv")
print(testDataTit.head())
```

# Checking % of *Men* and *Women* Survived

To check how many **females** and **males** *survived* during the titanic sink.

```python
womenSurvived = trainDataTit.loc[trainDataTit.Sex == 'female']["Survived"]
rateWomenSurvived = sum(womenSurvived)/len(womenSurvived)
print("Percentage of women who survived : ", rateWomenSurvived)
```

```python
menSurvived = trainDataTit.loc[trainDataTit.Sex == 'male']["Survived"]
rateMenSurvived = sum(menSurvived)/len(menSurvived)
print("Percentage of men who survived : ", rateMenSurvived)
```

**Conclusion :**
As we can see, *survival rate of WOMEN is **more** than MEN*, but it is not okay to say that all WOMEN survived and all MEN die.

# Selecting the Model :

Now, Applying ML Model - Here we use *Supervised Model* - (**Random Forest Model)**
* Splitting the data into training and testing sets
* Evaluatinig the Model to check accuracy score



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

y = trainDataTit["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"] # these features are taken to be in consideration for the prediction
x = pd.get_dummies(trainDataTit[features])
# x_test = pd.get_dummies(testDataTit[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=3)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

model.fit(x_train, y_train)
y_pred = model.predict(x_test)

```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred ))
```

```python
testDataTit.head()
X_test = pd.get_dummies(testDataTit[features])
```

```python
prediction=model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': testDataTit.PassengerId, 'Survived': prediction})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
